```python
from sklearn.feature_selection import chi2
import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer
# 经过卡方(chi2)检验后，找出了每个分类中关联度最强的两个词和两个词语对
N = 2
for cat in [0,1]:  
    tfidf = TfidfVectorizer(norm='l2', ngram_range=(1, 2))
    features = tfidf.fit_transform(data[data['cat_id']==cat].cut_review)
    labels = data[data['cat_id']==cat]['cat_id'].values.tolist()
    features_chi2 = chi2(features, labels)
    indices = np.argsort(features_chi2[0])
    feature_names = np.array(tfidf.get_feature_names())[indices]
    unigrams = [v for v in feature_names if len(v.split(' ')) == 1]
    bigrams = [v for v in feature_names if len(v.split(' ')) == 2]
    print("# '{}':".format(cat))
    print("  . Most correlated unigrams:\n       . {}".format('\n       . '.join(unigrams[-N:])))
    print("  . Most correlated bigrams:\n       . {}".format('\n       . '.join(bigrams[-N:])))
```

### 文本二分类算法封装

```python
import pandas as pd
import re
import random
import jieba
import joblib
import numpy as np
class Text:
    def __init__(self, data_path):
        self.data = pd.read_csv(data_path)
        
    def remove_punctuation(self, line):
        #定义删除除字母,数字，汉字以外的所有符号的函数
        line = str(line)
        if line.strip()=='':
            return ''
        #rule = re.compile(u"[^a-zA-Z0-9\u4E00-\u9FA5]")
        # 匹配除了汉字的所有字符
        rule = re.compile(u"[^\u4E00-\u9FA5]")
        line = rule.sub('',line)
        return line
    
    def stopwordslist(self, filepath):
        stopwords = [line.strip() for line in open(filepath, 'r', encoding='utf-8').readlines()]
        return stopwords
    
    def tokenization(self):
        # generate cut_review
        stopwords = self.stopwordslist('./stopWords')
        self.data['content_cleaning'] = self.data['content'].apply(self.remove_punctuation)
        self.data['cut_review'] = self.data['content_cleaning'].apply(lambda x: " ".join([w for w in list(jieba.cut(x)) if w not in stopwords]))
        
    def predict_text_class(self):
        # classify text with cut_review
        vec = joblib.load("CountVectorizer_trained_by_sample_train_set.tokenizer")
        classifier = joblib.load("MultinomialNB_model_trained_by_sample_train_set.model")
        x_data = self.data['cut_review'].apply(lambda x: np.str_(x)).values.tolist()
        #data_half['predicted_target'] = classifier.predict_proba(vec.transform(x_test_half))
        self.data['predicted_target'] = classifier.predict(vec.transform(x_data))
    
    def save_file(self, target_dir):
        # save file with predicted class
        self.data.to_csv(target_dir, index=False)
```
