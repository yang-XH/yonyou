### 更改num，为月份数量
* 输入数据格式应为 [ 服务，申请条数，服务，公告条数 ] *月份数量

```python
num = 3
name = []
```

```python
name += [['服务'+str(i)+'1','申请条数'+str(i),'服务'+str(i)+'2','公告条数'+str(i)] for i in range(1,num+1)]
name = [i for i in itertools.chain(*name)]
data.columns = name
```

```python
data
```

```python
# 合并各月份申请条数
df_app = pd.DataFrame(columns=['服务','申请条数']).set_index('服务')
for i in range(num):
    df = data[name[i*4:i*4+2]]
    df_app = df_app.join(df.set_index(name[i*4]),how='outer')
    df_app.dropna(axis=0,how='all',inplace=True)
    df_app.fillna(0, inplace=True)
    df_app.columns = ['申请条数','申请条数1']
    df_app['申请条数'] = df_app['申请条数'] + df_app['申请条数1']
    df_app.drop(columns=['申请条数1'],inplace=True)
```

```python
df_app.reset_index(inplace=True)
df_app.columns = ['上线申请','申请条数']
```

```python
df_app
```

```python
# 合并各月份公告条数
df_ann = pd.DataFrame(columns=['服务','公告条数']).set_index('服务')
for i in range(num):
    df = data[name[i*4+2:i*4+4]]
    df_ann = df_ann.join(df.set_index(name[i*4+2]),how='outer')
    df_ann.dropna(axis=0,how='all',inplace=True)
    df_ann.fillna(0, inplace=True)
    df_ann.columns = ['公告条数','公告条数1']
    df_ann['公告条数'] = df_ann['公告条数'] + df_ann['公告条数1']
    df_ann.drop(columns=['公告条数1'],inplace=True)
```

```python
df_ann.reset_index(inplace=True)
df_ann.columns = ['上线公告','公告条数']
```

```python
df_ann
```

```python
data = pd.concat([df_app,df_ann],axis=1)
```

## 桑基图处理代码

```python
data.columns = ['上线申请','申请条数','上线公告','公告条数']
# 赋值用loc[行，列]
data.loc[data['上线申请']=='CRM领域','上线申请'] = 'CRM云'
data.loc[data['上线公告']=='CRM领域','上线公告'] = 'CRM云'
a = data[['上线申请','申请条数']].copy()
b = data[['上线公告','公告条数']].copy()
a.dropna(axis=0,how='all',inplace=True)
b.dropna(axis=0,how='all',inplace=True)
m = pd.merge(a,b,left_on='上线申请',right_on='上线公告',how='outer')
```

```python
m['上线申请'] = m[['上线申请','上线公告']].apply(lambda x: x[0] if type(x[0])!=float else x[1],axis=1) # nan是float类型，if判断为True，需要用math.isnan判断
```

```python
m.fillna(0,inplace=True)
m['公告条数'] = m['公告条数'].astype('int')
m['上线公告'] = m['上线申请']
m['未提交条数'] = m['申请条数'] - m['公告条数']
```

```python
m
```

```python
data_normal = m
```

```python
k = data_normal[['上线申请','上线公告','公告条数']].copy()
k['上线公告'] = k['上线公告'].apply(lambda x:x+' - 已提交')
k.rename(columns={'公告条数':'value'},inplace=True)
k['value'] = k['value'].astype('int')
v = data_normal[['上线申请','上线公告','未提交条数']].copy()
v['上线公告'] = v['上线公告'].apply(lambda x:x+' - 待提交')
v.rename(columns={'未提交条数':'value'},inplace=True)
v['value'] = v['value'].astype('int')
data_final = pd.concat([k,v])
data_final.sort_values(by='上线申请',inplace=True)
data_final.reset_index(drop=True,inplace=True)
data_final['上线公告1'] = data_final['上线公告'] + "  " + data_final['value'].astype('str')
```

### 根据顺序排列

```python
ls = ['云平台', '应用平台', '数据中台',  '智能会计','财资云','企业绩效', '友报账', '电子会计档案', '税务云','人力云',  '协同云', '营销云', 'CRM云', '友云采','供应链', '制造云', '资产云']
data_final['ls'] = data_final['上线申请']
data_final['ls'] = data_final['ls'].astype('category')
```

```python
data_final['ls'].cat.reorder_categories(ls, inplace=True)
data_final.sort_values('ls',inplace=True)
data_final.reset_index(drop=True,inplace=True)
```

```python
nodes = []
# '上线申请'去重！
for i in data_final['上线申请'].unique().tolist()+data_final['上线公告1'].values.tolist():
    nodes.append({'name':i})
links = []
for i in data_final.values:
    links.append({'source':i[0],'target':i[3],'value':i[2]})
```

```python
from pyecharts.charts import Sankey
from pyecharts import options as opts
from pyecharts.globals import ThemeType
```

```python
c = (
    Sankey(init_opts=opts.InitOpts(width="800px", height="800px"))
    .add(
        "sankey",
        nodes,
        links,
        linestyle_opt=opts.LineStyleOpts(opacity=0.2, curve=0.5,color="source"),#, color="#ccebc5"
        label_opts=opts.LabelOpts(position="right"),
        itemstyle_opts=opts.ItemStyleOpts(border_width=1, border_color="#aaa"),
        tooltip_opts=opts.TooltipOpts(trigger_on="mousemove"),
        node_gap = 10,#节点之间的距离,(查看垂直图片的操作orient="vertical")
    )
    .set_global_opts(title_opts=opts.TitleOpts())#title="Sankey-基本示例"
    .render("sankey_base.html")
)
```
